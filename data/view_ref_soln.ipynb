```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading data
x = np.load('x.npy')

Nx = x.shape[0]

u_train = np.load('u_training_data.npy')
u_test  = np.load('u_test_data.npy')
Ntrain = u_train.shape[0]
Ntest  = u_test.shape[0]
```

```python
# Plotting a few training samples
Nview = 80

# Plotting a few training samples
fig,axs = plt.subplots(2,2,figsize=(10,10))
axs = axs.flatten()
for s in range (4):
    axs[s].plot(x[0:Nview],u_train[s,0:Nview],label='Measured component')
    axs[s].plot(x[Nview:],u_train[s,Nview:],label='Inferred component')
    axs[s].set_xlabel('x')
    axs[s].set_ylabel('u')
plt.savefig('samples.png',bbox_inches='tight')
```
